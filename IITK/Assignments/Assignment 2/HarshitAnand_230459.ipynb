```python
def outlier_treatment(df,feature):
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3-Q1
    lower_lim = Q1 - 1.5*IQR
    upper_lim = Q3 + 1.5*IQR
    return ((df[feature]<lower_lim) | (df[feature]>upper_lim))
     
```

### (iii) Train a multivariate linear regression model from scratch using numpy, pandas, and matplotlib. Additionally, train a linear regression model using the sklearn library on the training dataset. Also train ridge and lasso regression model using the sklearn library on the training dataset.

### 1. Manual Model

```python
def h(x,theta):
    return np.matmul(x,theta)
```

```python
def cost(x,theta,y):
    J = (h(x,theta)-y).T@(h(x,theta)-y)/(2*x.shape[0])
    # print("J shape = ",J)
    return J[0][0]
```

```python
def gradient_descent(iter,learning_rate,x,theta,y):
    J_all = [cost(x,theta,y)]
    for _ in range(iter):
        theta = theta - learning_rate*(x.T@(h(x,theta)-y))/x.shape[0]
        J_all.append(cost(x,theta,y))
    return theta,J_all
```

```python
def train(x,y,iter,learning_rate):
    theta = np.zeros((x.shape[1],1))
    parameters,J_all = gradient_descent(iter,learning_rate,x,theta,y)
    J_all = np.array(J_all).flatten()
    # print("Shape =",J_all.shape)
    # print("Shape =",np.linspace(1,(iter+1),(iter+1)).shape)
    plt.plot(np.linspace(1,(iter+1),(iter+1)),J_all)
    plt.title('Cost v/s Iterations')
    plt.xlabel("Iterations")
    plt.ylabel("Cost")
    return parameters
```

```python
x_training = x_train_norm.values
y_training = np.array(y_train['Y1']).reshape(-1,1)
x = np.hstack((np.ones((x_training.shape[0],1)),x_training))
parameters = train(x,y_training,100,0.1)
# x.shape
```

```python
parameters
```

```python
def performance_analysis(y,y_pred):
    SS_res = np.sum((y - y_pred)**2)
    SS_tot = np.sum((y - np.mean(y))**2)
    R2_score = 1-(SS_res/SS_tot)
    MSE = (y_pred-y).T@(y_pred-y)/(y.shape[0])
    RMSE = np.sqrt(MSE)
    return [R2_score, MSE[0][0], RMSE[0][0]]
```

```python
x_testing = np.hstack((np.ones((x_test_norm.shape[0],1)),np.array(x_test_norm)))
y_pred_scratch = np.matmul(x_testing,parameters)
```

```python
y_testing = np.array(y_test['Y1']).reshape(-1,1)
metrics_scratch= performance_analysis(y_testing,y_pred_scratch)
# y_testing.shape
print(metrics_scratch)
```
